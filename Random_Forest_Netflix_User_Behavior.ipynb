# Descriptive Statistics

We can use the describe() function to obtain summary statistics for the numerical features in the dataset. We use the describe() function to computes various summary statistics such as count, mean, standard deviation, minimum, first quartile, median (second quartile), third quartile, and maximum for given column. These statistics provide an overview of the distribution of data in each column of the dataset.

```python
numerical_stats = netflix_df.describe()
numerical_stats
```

## Categorical Features

Categorical data, or qualitative data, is essential non-numerical data.

By providing the 'object' argument to** describe()** we can view the unique values for each categorical column to understand the diversity in categorical variables better.

    describe(include=['object'])
    
Since we are dealing with categorical features count, unique, top, and frequency are provided.

The top and freq for each categorical variable give the most common value AND its frequency, offering insight into the data's skewness toward certain categories.

```python
# Set options to display all columns and increase display width
pd.set_option('display.max_columns', None)  # Ensure all columns are shown
pd.set_option('display.width', 300)        # Adjust width to your preference

# Get descriptive statistics for categorical columns
categorical_stats = netflix_df.describe(include=['object'])
print(categorical_stats)
```

# Correlation Analysis

```python
import numpy as np

# Convert 'churned' column to numerical format (0 for No, 1 for Yes)
netflix_df['churned_numeric'] = netflix_df['churned'].apply(lambda x: 1 if x == 'Yes' else 0)

# Calculating correlation matrix for numerical features, including the new 'churned_numeric' column
correlation_matrix = netflix_df.select_dtypes(include=np.number).corr() # Select only numerical columns for correlation calculation
churned_correlation = correlation_matrix['churned_numeric'].sort_values(ascending=False)
print(churned_correlation)
```
